# ***Chances of Admission Prediction***

```python
import pandas as pd
```

```python
admission = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
admission.head()
```

```python
admission.info()
```

```python
admission.describe()
```

```python
admission.columns
```

```python
y = admission['Chance of Admit ']
x = admission.drop(['Serial No', 'Chance of Admit '], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=2529)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
pred = model.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_percentage_error(y_test, pred)
```

```python
mean_absolute_error(y_test, pred)
```

```python
mean_squared_error(y_test, pred)
```

# ***Purchase Prediction And Micronumerosity***

```python
import pandas as pd
```

```python
purchase = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Customer%20Purchase.csv')
```

```python
purchase.head()
```

```python
purchase.info()
```

```python
purchase.describe()
```

```python
purchase.columns
```

```python
y = purchase['Purchased']
X = purchase.drop(['Purchased','Customer ID'],axis=1)
```

```python
X.replace({'Review':{'Poor':0,'Average':1,'Good':2}},inplace=True)
X.replace({'Education':{'School':0,'UG':1,'PG':2}},inplace=True)
X.replace({'Gender':{'Male': 0,'Female':1}},inplace=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
